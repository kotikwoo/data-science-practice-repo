```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import datasets
```

```python
X, y = datasets.make_regression(n_samples = 100, n_features = 1, noise = 20, random_state = 4)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1234)
```

```python
fig = plt.figure(figsize = (8, 6))
plt.scatter(X[:, 0], y, color = 'b', marker = 'o', s = 30)
plt.show()
```

Training the model

```python
learning_rate = 0.01
iterations = 1000
n_samples, n_features = X.shape
w = np.zeros(n_features)
b = 0
```

```python
for _ in range(iterations):
  y_predicted = np.dot(X_train, w) + b

  dw = (1/n_samples) * np.dot(X_train.T, (y_predicted - y_train))
  db = (1/n_samples) * np.sum(y_predicted - y_train)

  w = w - learning_rate * dw
  b = b - learning_rate * db
```

Predicting the values

```python
y_predicted = np.dot(X_test, w) + b
print(y_predicted)
```

Computing the mean squarred error(mse) between y and y_predicted

```python
mse = np.mean((y_test - y_predicted)**2)
print(mse)
```

Plotting the graph

```python
y_pred_line = np.dot(X, w) + b
cmap = plt.get_cmap('viridis')
fig = plt.figure(figsize = (8, 6))
m1 = plt.scatter(X_train, y_train, color = cmap(0.9), s = 10)
m2 = plt.scatter(X_test, y_test, color = cmap(0.5), s = 10)
plt.plot(X, y_pred_line, color = 'black', linewidth = 2, label = 'Prediction')
plt.show()
```
